```python
df = pd.read_csv('Dataset/kc_house_data.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df['Month Number'] = pd.to_datetime(df['date']).dt.month
```

```python
df['Month Name'] = pd.to_datetime(df['date']).dt.month_name()
```

```python
df
```

```python
df['Season'] = df['Month Number'].apply(lambda x: 'Spring' if 3 <= x <= 5 else 'Summer' if 6 <= x <= 8 else 'Fall' if 9 <= x <= 11 else 'Winter')
```
